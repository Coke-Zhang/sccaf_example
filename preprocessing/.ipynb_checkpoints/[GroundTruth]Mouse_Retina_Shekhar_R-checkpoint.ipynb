download from: ftp://ftp.ncbi.nlm.nih.gov/geo/series/GSE81nnn/GSE81904/suppl/GSE81904_bipolar_data_Cell2016.Rdata.gz

```r
load("GSE81904_bipolar_data_Cell2016.Rdata")
print(dim(bipolar_dge))
```

```r
mt.genes = grep("mt-", rownames(bipolar_dge), value = TRUE)
cells.use = colnames(bipolar_dge)[colSums(bipolar_dge[mt.genes, ])/colSums(bipolar_dge) <
0.1]
bipolar_dge = bipolar_dge[, cells.use]
dim(bipolar_dge)
```

```r
#Execution time ~ 10-15 minutes
dsq.bip=scDrop(count.data=bipolar_dge)
dsq.bip=initialize(dsq.bip, min.genes = 500, min.cells = 30, min.counts=60)
```

```r
print(dim(dsq.bip@data))
```

```r
dsq.bip = readRDS("Shekhar.RDS")
```

```r
table(dsq.bip@meta$sample)
```

```r
violinplot(dsq.bip,c("num.genes","num.trans"))
```

```r
batchname = as.character(dsq.bip@meta$sample)
batchid = rep(1,length(batchname))
batchid[batchname=="Bipolar5"] = 2
batchid[batchname=="Bipolar6"] = 2
```

```r
class(dsq.bip@data)
```

```r
typeof(dsq.bip@data)
```

```r
as.matrix(dsq.bip@data)[1:5,1:5]
```

```r
correct.data = ComBat(as.matrix(dsq.bip@data),batchid, prior.plots=FALSE, par.prior=TRUE)
```

```r
max.val=6
correct.data[correct.data > max.val] = max.val
            
dsq.bip@batch.data = as.data.frame(correct.data)
rm(correct.data)
dsq.bip@scale.data = t(scale(t(dsq.bip@batch.data), center=TRUE, scale=TRUE))
```

save the corrected data as a table

```r
write.csv(dsq.bip@scale.data, "Shekhar_corrected.csv")
```
