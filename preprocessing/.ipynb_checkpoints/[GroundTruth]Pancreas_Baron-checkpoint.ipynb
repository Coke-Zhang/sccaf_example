## Human

> Preprocessing

```python
from glob import iglob
ctss=[]
for f in iglob("GSM*_human*_umifm_counts.csv.gz"):
    cts= pd.read_csv(f, sep=',',compression='gzip',index_col=0)
    ctss.append(cts)
```

```python
cts = pd.concat(ctss, axis=0)
```

```python
cts.shape
```

```python
cts.head()
```

```python
obs = cts.iloc[:,:2]
```

```python
obs.head()
```

```python
cts = cts.iloc[:,2:]
```

```python
cts.head()
```

```python
ad = sc.AnnData(X=scipy.sparse.csr_matrix(cts.values))
ad.obs_names = cts.index
ad.var_names = cts.columns
for col in obs.columns:
    ad.obs[col] = obs[col]
```

```python
ad.raw = sc.pp.log1p(ad,copy=True)
```

```python
ad.shape
```

```python
ad.obs['lib'] = ad.obs_names.str.split('.').str[0]
ad.obs['lib'] = ad.obs['lib'].astype('category')

ad.obs['donor'] = ad.obs_names.str.split('_').str[0]
ad.obs['donor'] = ad.obs['donor'].astype('category')
```

```python
ad.obs['cell'] = ad.obs['assigned_cluster']
ad.obs['cell'].value_counts()
```

```python
ad.write("../write/Baron_Human.h5")
```

```python
ad = sc.read("../write/Baron_Human.h5")
```

```python
ad.X
```

```python
ad.raw.X
```

```python
# ad = ad[~ad.obs['cell'].isin(['t_cell','mast','epsilon','schwann']),:]
sc.pp.filter_cells(ad, min_genes=200)
sc.pp.filter_cells(ad, max_counts=12000)
sc.pp.filter_genes(ad, min_cells=3)
```

```python
ad = ad[:,:-1]
```

```python
ad.shape
```

```python
sc.pp.normalize_per_cell(ad, counts_per_cell_after=1e4)
filter_result = sc.pp.filter_genes_dispersion(
    ad.X, min_mean=0.05, max_mean=10, min_disp=0.25)
sc.pl.filter_genes_dispersion(filter_result)
print("n_HVGs: %d"%sum(filter_result.gene_subset))
ad = ad[:, filter_result.gene_subset]
```

```python
sc.pp.log1p(ad)
# use NaiveDE to regress out the technical effects.
import NaiveDE
res = NaiveDE.regress_out(ad.obs, ad.X.T, 'n_genes+n_counts+C(donor)', rcond=1e-4).T
ad.X = res

sc.pp.scale(ad, max_value=10)
sc.tl.pca(ad)
sc.tl.tsne(ad, n_pcs=30)
```

```python
figsize(5,5)
sc.pl.tsne(ad, color=['cell','donor'])
```

```python
sc.pp.neighbors(ad, n_pcs=25)
sc.tl.louvain(ad, resolution=1.5)
```

## Mouse

```python
from glob import iglob
ctss=[]
for f in iglob("GSM*_mouse*_umifm_counts.csv.gz"):
    print(f)
    cts= pd.read_csv(f, sep=',',compression='gzip',index_col=0)
    ctss.append(cts)
```

```python
cts = pd.concat(ctss, axis=0)
```

```python
cts.shape
```

```python
cts.head()
```

```python
obs = cts.iloc[:,:2]
```

```python
obs.head()
```

```python
cts = cts.iloc[:,2:]
```

```python
cts.head()
```

```python
ad = sc.AnnData(X=scipy.sparse.csr_matrix(cts.values).tocsc())
ad.obs_names = cts.index
ad.var_names = cts.columns
for col in obs.columns:
    ad.obs[col] = obs[col]
```

```python
ad.raw = sc.pp.log1p(ad,copy=True)
```

```python
ad.shape
```

```python
ad.obs['lib'] = ad.obs_names.str.split('.').str[0]
ad.obs['lib'] = ad.obs['lib'].astype('category')

ad.obs['donor'] = ad.obs_names.str.split('_').str[0]
ad.obs['donor'] = ad.obs['donor'].astype('category')

ad.obs['cell'] = ad.obs['assigned_cluster']
```

```python
ad.write("../write/Baron_Mouse.h5")
```

```python
ad = sc.read("../write/Baron_Mouse.h5")
```

```python
ad.X
```

```python
ad.raw.X
```

```python
sc.pp.filter_cells(ad, min_genes=500)
sc.pp.filter_cells(ad, min_counts=500)
sc.pp.filter_genes(ad, min_cells=3)
```

```python
ad.obs['assigned_cluster'].value_counts()
```

```python
sc.pp.normalize_per_cell(ad, counts_per_cell_after=1e4)
filter_result = sc.pp.filter_genes_dispersion(
    ad.X, min_mean=0.05, max_mean=10, min_disp=0.25)
sc.pl.filter_genes_dispersion(filter_result)
print("n_HVGs: %d"%sum(filter_result.gene_subset))
ad = ad[:, filter_result.gene_subset]
```

```python
sc.pp.log1p(ad)
ad.obs['log_n_counts'] = np.log(ad.obs['n_counts'])
ad.obs['log_n_genes'] = np.log(ad.obs['n_genes'])

sc_pp_regress_out(ad, keys=['log_n_genes','lib'])

sc.tl.pca(ad)
sc.tl.tsne(ad)
```

```python
sc.pl.tsne(ad, color=['cell','donor','lib'])
```

```python
sc.pp.neighbors(ad)
# sc.tl.umap(ad)
sc.tl.louvain(ad, resolution=1.5)
```
