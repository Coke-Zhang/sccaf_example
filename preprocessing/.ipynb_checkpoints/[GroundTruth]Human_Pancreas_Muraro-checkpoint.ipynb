```python
%pylab inline

import warnings
warnings.filterwarnings("ignore")

from SCCAF import *
```

download counts

```python
cts = pd.read_csv("ftp://ftp.ncbi.nlm.nih.gov/geo/series/GSE85nnn/GSE85241/suppl/GSE85241_cellsystems_dataset_4donors_updated.csv.gz",\
                 sep='\t', index_col=0)
```

```python
cts.head()
```

```python
cts.shape
```

```python
obs = pd.read_csv("https://s3.amazonaws.com/scrnaseq-public-datasets/manual-data/muraro/cell_type_annotation_Cels2016.csv",\
                  sep='\t', index_col=0)
obs.index = obs.index.str.replace('.','-')
obs.columns = ['cell']
```

```python
ad = sc.AnnData(scipy.sparse.csr_matrix(cts.values.T))
ad.obs_names = cts.columns
ad.var_names = cts.index
ad = ad[ad.obs_names.isin(obs.index),:]
ad.obs = obs.loc[ad.obs_names]
```

```python
ad.raw = sc.pp.log1p(ad,copy=True)
```

```python
ad.shape
```

```python
ad.write("../write/Muraro2016Pancreas.h5")
```

```python
ad = sc.read("../write/Muraro2016Pancreas.h5")
```

```python
ad.X
```

```python
ad.raw.X
```

```python
sc.pp.filter_cells(ad, min_genes=200)
sc.pp.filter_genes(ad, min_cells=3)
```

```python
ad = ad[~ad.obs['cell'].isin(['nan','unclear']),:]
```

```python
ad.obs['n_counts'] = ad.X.sum(axis=1)
ad.obs['n_genes'] = (ad.X>0).sum(axis=1)
ad.var_names_make_unique()
```

```python
sc.pp.normalize_per_cell(ad, counts_per_cell_after=1e4)
```

```python
sc.pp.log1p(ad)
# sc_pp_regress_out(ad,keys=['n_counts','n_genes','individual'])
# sc.pp.scale(ad, max_value=10)
sc.tl.pca(ad)
```

```python
sc.tl.tsne(ad, n_pcs=15)
sc.pp.neighbors(ad, n_pcs=15)
sc.tl.umap(ad)
sc.tl.louvain(ad, resolution=0.5)
```

```python
figsize(5,5)
sc.pl.tsne(ad, color=['cell','louvain','n_counts'])
```

```python
ad.write("../write/Muraro2016Pancreas_reg.h5")
```
