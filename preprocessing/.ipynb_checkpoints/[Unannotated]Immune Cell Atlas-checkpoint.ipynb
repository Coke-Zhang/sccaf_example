```python
adc = readHCA('raw_data/ica_cord_blood.h5')
adb = readHCA('raw_data/ica_bone_marrow.h5')
```

```python
adc.obs['tissue'] = 'ABM'
adb.obs['tissue'] = 'ACB'
```

```python
ad = adc.concatenate(adb)
ad.var_names_make_unique()
```

```python
ad.write('data/ica.h5')
```

```python
ad = sc.read('data/ica.h5')
```

```python
ad.shape
```

```python
ad.raw = sc.pp.log1p(ad,copy=True)
```

```python
sc.pp.filter_cells(ad, min_genes=200)
sc.pp.filter_genes(ad, min_cells=2)
```

```python
mito_genes = ad.var_names.str.startswith('MT-')
# for each cell compute fraction of counts in mito genes vs. all genes
# the `.A1` is only necessary as X is sparse (to transform to a dense array after summing)
ad.obs['percent_mito'] = np.sum(
    ad[:, mito_genes].X, axis=1).A1 / np.sum(ad.X, axis=1).A1
# add the total counts per cell as observations-annotation to adata
ad.obs['n_counts'] = ad.X.sum(axis=1).A1
```

```python
sc.pp.normalize_per_cell(ad, counts_per_cell_after=1e4)
sc.pp.log1p(ad)
```

```python
sc.pp.highly_variable_genes(ad, flavor='cell_ranger', n_top_genes=2000)
print('\n','Number of highly variable genes: {:d}'.format(np.sum(ad.var['highly_variable'])))
```

```python
sc.tl.pca(ad)
```

```python
%time
sc.pp.neighbors(ad)
sc.tl.umap(ad)
```

```python
sc.tl.louvain(ad)
```

```python
figsize(5,5)
sc.pl.umap(ad,color=['louvain','CD34','SPINK2'])
```

```python
figsize(5,5)
sc.pl.umap(ad,color=['tissue','louvain'])
```

```python
ad.write("processed/ica_all.h5")
```

```python
sc.pp.highly_variable_genes(ad, flavor='cell_ranger', n_top_genes=2000)
print('\n','Number of highly variable genes: {:d}'.format(np.sum(ad.var['highly_variable'])))
```

```python
import NaiveDE
ebat = NaiveDE.regress_out(ad.obs, ad[:,ad.var['highly_variable']].X.T.todense(),\
                           'C(batch)+np.log(n_counts)',rcond=1e-4).T
```

```python
ebat.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(50, whiten=True)
ad.obsm['X_pca'] = pca.fit_transform(ebat)
sc.pp.neighbors(ad)
sc.tl.umap(ad)
```

```python
ad.uns['batch_colors'] = default_20
```

```python
figsize(5,5)
sc.pl.umap(ad,color=['batch','CD34','SPINK2','louvain'])
```

```python
sc.tl.louvain(ad)
```

```python
figsize(5,5)
sc.pl.umap(ad,color=['batch','CD34','SPINK2','louvain'])
```

```python
figsize(5,5)
sc.pl.umap(ad,color=['tissue'])
```
