## ACC

```python
cts = pd.read_csv("human_brain/human_ACC_2018-10-04_exon-matrix.csv", index_col=0)
var = pd.read_csv("human_brain/human_ACC_2018-10-04_genes-rows.csv", index_col=0)
obs = pd.read_csv("human_brain/human_ACC_2018-10-04_samples-columns.csv", index_col=0)
ins = pd.read_csv("human_brain/human_ACC_2018-10-04_intron-matrix.csv", index_col=0)
```

```python
cc = scipy.sparse.csr_matrix(cts.T)
ii = scipy.sparse.csr_matrix(ins.T)
exp = cc+ii
ad = sc.AnnData(exp)
ad.obs = obs
ad.var = var
ad.layers['intron'] = ii
ad.layers['exon'] = cc
```

```python
mkdir processed
```

```python
ad.write("processed/ACC.h5")
```

## combine

```python
dfs = []
for i in ['ACC','LGN','MTG','VISp']:
    ad = sc.read("processed/%s.h5"%i)
    ad.obs['tissue'] = i
    dfs.append(ad)
ad = dfs[0].concatenate(tuple(dfs[1:]))
```

```python
ad.write('processed/all.h5')
```

# Analysis

```python
ad = sc.read('processed/all.h5')
```

```python
ad.shape
```

```python
sc.pp.filter_cells(ad, min_genes=200)
sc.pp.filter_cells(ad, min_counts=200)
sc.pp.filter_genes(ad, min_cells=2)
sc.pp.normalize_per_cell(ad, counts_per_cell_after=1e4)
```

```python
ad.shape
```

```python
sc.pp.log1p(ad)
ad.raw = ad
```

```python
sc.pp.highly_variable_genes(ad, flavor='cell_ranger', n_top_genes=2000)
print('\n','Number of highly variable genes: {:d}'.format(np.sum(ad.var['highly_variable'])))
```

```python
sc.tl.pca(ad, svd_solver='arpack')
sc.pp.neighbors(ad)
sc.tl.umap(ad)
sc.tl.louvain(ad)
```
