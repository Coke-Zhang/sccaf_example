```python
df
```

```python
import time
import datetime
```

```python
xs = []
for s in df['time on raw']:
    if s.find('min')>-1:
        x = time.strptime(s,'%Mmin %Ss')
        x = datetime.timedelta(minutes=x.tm_min,seconds=x.tm_sec).total_seconds()
    elif s.endswith('ms'):
        x = float(s.replace('ms',''))/1000
    else:
        x = float(s.replace('s',''))
    xs.append(x)
```

```python
df['time on raw'] = xs
```

```python
xs = []
for s in df['time on PCA']:
    if s.find('min')>-1:
        x = time.strptime(s,'%Mmin %Ss')
        x = datetime.timedelta(minutes=x.tm_min,seconds=x.tm_sec).total_seconds()
    elif s.endswith('ms'):
        x = float(s.replace('ms',''))/1000
    else:
        x = float(s.replace('s',''))
    xs.append(x)
```

```python
df['time on PCA'] = xs
```
